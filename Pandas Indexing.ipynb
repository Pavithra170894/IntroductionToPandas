```python
import pandas as pd
```

```python
# Two types of indexing
# loc - label-based selection
# iloc - index-based selection
```

```python
exam_scores = pd.read_csv("exam_scores.csv")
```

```python
exam_scores.head()
```

```python
# INDEX BASED SELECTION

# df[rows, columns]
# first row, first column
exam_scores.iloc[0,0]
```

```python
# rows in range, first column
exam_scores.iloc[:10,2:]
```

```python
# negative indexes
exam_scores.iloc[-5:,2:3]
```

```python
# LABEL BASED SELECTION
```

```python
# loc[rows, columns]
exam_scores.loc[:, ["math score", "lunch"]]
```
